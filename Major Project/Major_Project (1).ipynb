```python
from nltk.corpus import stopwords 
import string
import re

#function to preprocess text

pattern = "[^a-zA-Z]"
def text_preprocessing(mess):
    nopunc=[char for char in mess if char not in string.punctuation] #removing punctuations
    nopunc=''.join(nopunc)
    nopunc = re.sub(pattern," ",nopunc)
    return [word for word in nopunc.split() if word.lower() not in stopwords.words('english')] # removing stopwords
```

```python
#function to calculate occurances of words
def most_frequent(List): 
    LIST=set(List)
    dict={}
    for i in LIST:
        count=List.count(i)
        dict[i]=count
    return dict
```

```python
L_Male=list()
for i in Male.index:
    L_Male+=Male[i]
male_dict=most_frequent(L_Male)
male_dict
```

# Data Preprocessing , Feature Selection

```python
#preprocessing of the data to be trained and tested with

df_final=pd.concat([df1,df2]) #selection of rows with target values as male and female only

df_final=df_final[['gender','description','text']]

df_final=df_final.dropna() #dropping rows with null values
df_final = df_final.reset_index(drop = True) # re indexing the rows

df_final['desc+text']=df_final['description']+df_final['text'] #combing the text in description and tweet into single one
df_final['desc+text']
```

```python
df_final.columns
```

```python
# the coloumn with combined description and tweet 

df_final['desc+text']
```

```python
# appying the function created earlier to preprocess the data

Text=df_final['desc+text'].apply(text_preprocessing)
```

```python
Text
```

```python
# funtion for converting to lower case
def tolower(case):
    for i in range(len(case)):
        case[i]=case[i].lower()
    return case
```

```python
# funcation for reomving typos and meaning less words

def typos(miss):
    misspelled = list(spell.unknown(miss))
    for i in misspelled:
        if i in miss:
            miss.remove(i)
    return miss
```

```python
#converting to lower case

Text_new=Text.apply(tolower)
Text_new
```

```python
#removing typos

Text_New=Text_new.apply(typos)
Text_New
```

```python
from nltk.stem import WordNetLemmatizer 
lem = WordNetLemmatizer() 

#function to lemmatize the words

def lemmatizer(List):
    for i in range(len(List)):
        List[i]=lem.lemmatize(List[i])
    return List
```

```python
# lemmatizing the data

Text_New=Text_New.apply(lemmatizer)
Text_New
```

```python
L_Text_New=list()
for i in Text_New.index:
    L_Text_New+=Text_New[i]
Text_New_dict=most_frequent(L_Text_New)
len(Text_New_dict)
```

```python
#counting the most commonly used words

count_text=Counter(Text_New_dict)
text_com= count_text.most_common(100)
Text_vocab=list()
Text_count=list()
for i in text_com: 
    Text_vocab.append(str(i[0]))
    Text_count.append(int(i[1]))
    
print(Text_vocab[:5],Text_count[:5])
```

# Ensemble model using bagging technique

```python
from sklearn.model_selection import train_test_split 

#creating a training set of 80% and a testing set of 20%

x_train,x_test,y_train,y_test=train_test_split(sdf,df_final['gender'],test_size=0.2)
x_train.shape
```

```python
#creating 3 empty sets of dataframe for each model for x and y 

model1x=pd.DataFrame()
model2x=pd.DataFrame()
model3x=pd.DataFrame()
model1y=pd.DataFrame()
model2y=pd.DataFrame()
model3y=pd.DataFrame()
```

```python
#spliting the traing set in 3 sets of 33% each and feeding int into each data frames created above
# repeating these three times so that no of rows in the 3 dataframes are same as the original training set of 80%

for i in range(3):
    X1,X2,Y1,Y2=train_test_split(x_train,y_train,test_size=0.33)
    X11,X22,Y11,Y22=train_test_split(X1,Y1,test_size=0.5)
    
    model1x=pd.concat([model1x,X2])
    model1y=pd.concat([model1y,Y2])
    
    model2x=pd.concat([model2x,X11])
    model2y=pd.concat([model2y,Y11])
    
    model3x=pd.concat([model3x,X22])
    model3y=pd.concat([model3y,Y22])

    
print(model1y.shape,model2x.shape,model3x.shape) #priting the shape of the the training sets
```

```python
#inputing the models on which the data is to be trained

from sklearn.linear_model import LogisticRegression

from sklearn.tree import DecisionTreeRegressor

from sklearn.naive_bayes import MultinomialNB

from sklearn.neighbors import KNeighborsClassifier

neigh = KNeighborsClassifier(n_neighbors=5)
lm=LogisticRegression()
dtree=DecisionTreeRegressor()
mnb=MultinomialNB()
```

```python
#fitting the models with the training data

lm.fit(model1x,model1y)
dtree.fit(model2x,model2y)
mnb.fit(model3x,model3y)
neigh.fit(x_train,y_train)
```

```python
#predicting the output from the testing data of 20% created earlier

pred1=lm.predict(x_test)
pred2=dtree.predict(x_test)
pred3=mnb.predict(x_test)
pred4=neigh.predict(x_test)
```

```python
#calculating the accuracy scores of individual models

from sklearn.metrics import accuracy_score

Acc=accuracy_score(y_test,pred1)

Acc2=accuracy_score(y_test,pred2.round())

Acc3=accuracy_score(y_test,pred3)

Acc4=accuracy_score(y_test,pred4)

print(Acc,Acc2,Acc3,Acc4)
```

```python
#calculating the prediction by casting a vote on the output of each model (part of bagging technique) 

pred_final=[]
from collections import Counter 
for i in range(len(pred1)):
    l=[]
    l.append(pred1[i])
    l.append(pred2[i])
    l.append(pred3[i])
    l.append(pred4[i])
    count=Counter(l)
    pred_final.append(count.most_common(1)[0][0])
```

```python
#predicting accuracy of the ensemble model

Acc_final=accuracy_score(y_test,pred_final)
Acc_final
```
