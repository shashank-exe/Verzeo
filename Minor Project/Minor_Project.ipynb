# 1) Which are the movies with the third-lowest and third-highest budget?

```python
#Sorting the dataset in ascending order of budget
df = df.sort_values(["budget"])
#Resetting the index of the sorted dataset
df = df.reset_index(drop = True)
#Printing the title & budget of the movie with the 3rd lowest budget
print('Title & budget of the movie with the 3rd lowest budget')
df.iloc[[2],[5,3]]  #Indices of the 'original_title' & 'budget' columns are 5 & 3 respectively
```

```python
#Sorting the dataset in descending order of budget
df = df.sort_values(["budget"], ascending=[False])
#Resetting the index of the sorted dataset
df = df.reset_index(drop = True)
#Printing the title & budget  of the movie with the 3rd highest budget
print('Title & budget of the movie with the 3rd highest budget')
df.iloc[[2],[5,3]]
```

# 2) What is the average number of words in movie titles between the years 2000-2005?

```python
#Subsetting the movies which were released between 2000 & 2005
df2 = df[(df["release_year"] >= 2000) & (df["release_year"] <= 2005)]
df2.head()
```

```python
print(f'Total Number of Movies: {df.shape}')
print(f'Number of Movies between 2000-2005: {df2.shape}')
```

```python
#Creating a new column 'word_count' which stores the number of words in 'original_title'
df2['word_count'] = df2['original_title'].str.split().str.len()
df2[["original_title","word_count"]]
```

```python
#Storing 'word_count' in an array 'count'
count = np.array(df2['word_count'])
#Printing the mean of 'count' array
print(f'The average number of words in movie titles between the years 2000-2005: {np.mean(count)}')
```

# 4) Which are the movies with the most and least earned revenue?

```python
#Removing the movies whose revenues are less than 100
df = df[df["revenue"] > 100]
```

```python
#Printing the row of the movie with the highest revenue
highest_rev = df.original_title[df.revenue == df.revenue.max()]
print(f'The movie with the highest revenue: {highest_rev}')
```

```python
#Printing the row of the movie with the least revenue
lowest_rev = df.original_title[df.revenue == df.revenue.min()]
print(f'The movie with the lowest revenue: {lowest_rev}')
```
